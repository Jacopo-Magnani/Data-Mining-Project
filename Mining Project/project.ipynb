```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import ScalarFormatter
import seaborn as sn
```

```python
#_PFAS
```

```python
#Download and print out PFAS file
PFAS_df = pd.read_csv('PFAS_J.csv')

print(PFAS_df.shape)
```

```python
#Clean columns
PFAS_df.drop(columns = ['Unnamed: 0', 'SEQN', 'WTSB2YR'], inplace = True)

#Clean rows if less than 50%
count = PFAS_df.isna().sum(axis = 1)
threshold = PFAS_df.shape[1] * 0.5
PFAS_df = PFAS_df[count <= threshold]

zero_tf = (PFAS_df > -0.0009) & (PFAS_df < 0.0009)
zero_count = zero_tf.sum(axis = 0)
tf1 = zero_count >= threshold
PFAS_df.drop(columns = tf1[tf1].index, inplace = True)
print(PFAS_df)
print(PFAS_df.shape)
```

```python
#Correlation Matrix
corr_matrix = PFAS_df.corr()
print(corr_matrix.shape)
```

```python
#Heatmap
hm = sn.heatmap(data = corr_matrix)
```

```python
#High Correlation Heatmap and Scatter Plots
tf = (corr_matrix > 0.85) & (corr_matrix < 1.0)
pearson_correlation_filtered_data = corr_matrix[tf]

hm = sn.heatmap(data = pearson_correlation_filtered_data, annot = True, annot_kws={"color": "black"})
plt.title('Filtered Correlation Heatmap')
plt.show()

pairs = []
for i in range(len(pearson_correlation_filtered_data)):
    for j in range(i + 1, len(pearson_correlation_filtered_data)):
        if not pearson_correlation_filtered_data.iloc[i, j] >= 0.85:
            continue
        pair = (pearson_correlation_filtered_data.columns[i], pearson_correlation_filtered_data.columns[j])
        pairs.append(pair)
        plt.scatter(PFAS_df[pair[0]], PFAS_df[pair[1]])
        plt.xlabel(pair[0])
        plt.ylabel(pair[1])
        plt.title(f'Scatter Plot of {pair[0]} vs {pair[1]}: {pearson_correlation_filtered_data.iloc[i, j]:.2f}')
        plt.show()
```
